```python
import cv2
import os
import pandas as pd
from recognizer import PlateRecognizer
from pipeline import LicensePlatePipeline
```

```python
# Setup paths
project_path = os.path.dirname(os.getcwd())
api_path = os.path.join(project_path, 'api')

# Checkpoint paths
yolo_ckpt = os.path.join(project_path, 'checkpoints', 'detect_best.pt')
digit_ckpt = os.path.join(project_path, 'checkpoints', 'digit_best.pth')
letter_ckpt = os.path.join(project_path, 'checkpoints', 'letter_best.pth')

# Initialize model
model = PlateRecognizer(
    yolo_ckpt=yolo_ckpt,
    digit_ckpt=digit_ckpt,
    letter_ckpt=letter_ckpt
)

# Initialize pipeline
pipeline = LicensePlatePipeline(model, api_path)

print("✓ System initialized and ready")
```

```python
# Video paths
video_in_path1 = os.path.join(project_path, "videos", "input1.mp4")
video_in_path2 = os.path.join(project_path, "videos", "input2.mp4")

video_out_path1 = os.path.join(project_path, "videos", "output1.mp4")
video_out_path2 = os.path.join(project_path, "videos", "output2.mp4")

print(f"Input video: {video_in_path2}")
print(f"Output video: {video_out_path2}")
```

# Process Videos and Save Detections

```python
# Process video and save detections to database
result = pipeline.process_video(
    video_in_path2,
    skip_frames=2,
    vote_frames=5,
    similarity_threshold=0.85,
    min_confidence=0.75,
    save_to_db=True,
    verbose=True
)
```

# Create Visualization (Optional)

```python
# Create annotated video with plate detections
pipeline.visualize_video(video_in_path2, video_out_path2)
```

```python
# Query and display recent detections
detections = pipeline.get_recent_detections(limit=20)
df = pd.DataFrame(detections)

print("\n📊 Recent Detections from Database:")
print(df.to_string(index=False))
print(f"\nTotal detections: {len(detections)}")
```
